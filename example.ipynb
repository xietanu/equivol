```python
import cv2
from matplotlib import pyplot as plt

from ev_segment import ev_segment

%load_ext autoreload
%autoreload 2
```

```python
EXAMPLE = ("1.png")

mask = cv2.imread(f"example_masks/{EXAMPLE}", cv2.IMREAD_UNCHANGED)

vol_mask = ev_segment(
    mask=mask,
    layer_thicknesses=[1,1,2,0.6,1,1],
    scale=2,
    smoothing_degree=50,
    report=True
)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].imshow(mask, cmap="plasma")
ax[0].set_title("Input Mask")
ax[1].imshow(vol_mask, cmap="plasma")
ax[1].set_title("EV Segmented Mask")
plt.show()
```
